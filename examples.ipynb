# This notebook demonstrates how to use these scripts

```python
# first, you'll want to import main from wherever this script lives
import main
```

```python
# use this function to see what exchanges are available in this code
# the goal is to add more exchanges later on
main.get_valid_exchanges()
```

```python
# use this function to get information on all trading pairs from the exchange
# beware that not all trading pairs returned are currently active
main.get_trading_pairs('coinbase')
```

```python
# use the candles class to pull candlestick data from the exchange
# first we'll grab daily BTC-USD data
btc_daily = main.Candles(
    exchange='coinbase',
    ticker_id='BTC-USD',
    start_date='2023-01-01',
    end_date='2023-01-03'
)
```

```python
# access the df like this
btc_daily.df
```

```python
# we can do the same thing for ETH-USD
eth_daily = main.Candles(
    exchange='coinbase',
    ticker_id='ETH-USD',
    start_date='2023-01-01',
    end_date='2023-01-03'
)
```

```python
eth_daily.df
```

```python
# we can pull data for smaller time frames as well such as the hourly time frame
# the documentation provides which time intervals are available
btc_hourly = main.Candles(
    exchange='coinbase',
    ticker_id='BTC-USD',
    start_date='2023-01-01',
    end_date='2023-01-01',
    start_time='10:00:00',
    end_time='14:00:00', # end dates and times are inlusive!
    local_timezone='America/New_York',
    time_interval='1h'
)
```

```python
btc_hourly.df
```
